The boston dataset is a collection of descriptive statistics that relate to house prices in the boston area from 1970. The statics were colleted by the US Census Service in the 1970 census of population and housing. The samples show summary statisctics for cenus tracts (areas) within the Boston metropolitan statistical area(MSA). The data is taken from a paper that looks at the effect of air polution on house prices in the boston area from 1978 "Hedonic housing prices and the demand for clean air".
There were 506 individual cenus tracts in the Boston MSA in 1970. Today there are many more due to the expansion of the metropolitan area since 1970. Each tract is represented by a row below in the database.

After some research into the 1970 cenus I was able to find the questions asked. The price question goes up in ranges and limits itself to 50,000 or more. So for the price varible to be 50.00 every house hosue within that area must be over 50,000 dollars meaning that evry data point at 50.00 must be above that.

<img src="img/priceq.png" alt="Image of boston" >

Because of this the dependant variable Price is skewed. There are 3 options here remove the data points at 50.00, leave the data points at 50.00 in the dataset  or redistrobute the data points at 50.00 in a form that mathes the distrobution of the dataset better. In this case it is best to redistrobute the datapoints at 50.00. It is not the perfect answer but leaving the points were they are or deleting them skews the dataset even more than retdistrobuting them.

```python
## The descriptive statistics of the price variable
df['PRICE'].describe()
```

```python
## importing skewnorm a repositry that creates skewed normal distrobutions under certain parameters
from scipy.stats import skewnorm
from scipy import stats
import matplotlib.pyplot as plt
```

```python
## a is the amount to skewe the data , loc is the mean value this has been increased because the paremters above
## 50 will change this value in the dataset, scale is standard deviations this has also been increased to account 
## for teh increase above 50.00
a, loc, scale  = 3.3, 24.25,  12.15
# draw a sample
data = stats.skewnorm(a, loc, scale).rvs(506)
# estimate parameters from sample
ae, loce, scalee = stats.skewnorm.fit(data)
```

```python
## distplot of the new variables
sns.distplot(data,bins=100)
```

```python
## sorting the new dataset by size
data.sort()
```

```python
## chnaging that sort from accending to decending
data =data[::-1] 
```

```python
## only taking the top 16 variables to replace the 16 variables at 50.00
data = data[:16]
```

```python
## printing the replacemnt variables
data
```

## with k folds

```python
xwhite_train, xwhite_test, outputs_train, outputs_test = mod.train_test_split(xwhite, outputs, test_size=110)
```

```python
### modified from https://www.kaggle.com/shanekonaung/boston-housing-price-dataset-with-keras


from keras import models
from keras import layers

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(128, activation='relu', input_dim=4))
    model.add(layers.Dense(128, activation='sigmoid'))
    model.add(layers.Dense(1 ,activation='linear'))

    model.compile(optimizer='adam',
              loss='mse',
              metrics=['mae'])
    return model
```

```python
### modified from https://www.kaggle.com/shanekonaung/boston-housing-price-dataset-with-keras


k = 5
num_val_samples = len(xwhite_train) // k
num_epochs = 400
all_scores = []

for i in range(k):
    print(f'Processing fold # {i}')
    val_data = xwhite_train[i * num_val_samples: (i+1) * num_val_samples]
    val_targets = outputs_train[i * num_val_samples: (i+1) * num_val_samples]
    
    partial_xwhite_train = np.concatenate(
                            [xwhite_train[:i * num_val_samples],
                            xwhite_train[(i+1) * num_val_samples:]],
                            axis=0)
    partial_outputs_train = np.concatenate(
                            [outputs_train[:i * num_val_samples],
                            outputs_train[(i+1)*num_val_samples:]],
                            axis=0)
    model = build_model()
    model.fit(partial_xwhite_train,
              partial_outputs_train,
              epochs=num_epochs,
              batch_size=8,
              verbose=0)
    val_mse, val_mae = model.evaluate(val_data, val_targets, verbose=1)
    all_scores.append(val_mae)
```

```python
print(f'all_scores : {all_scores}')
print(f'mean all scores : {np.mean(all_scores)}')
```
